```python
import os 
import pandas as pd 
```

```python
# save path to data set in a variable 
# use pandas to read data 
schools_complete = "schools_complete.csv"
students_complete = "students_complete.csv"

schools_complete_df = pd.read_csv(schools_complete)
students_complete_df = pd.read_csv(students_complete)
```

```python
#schools DF 
schools_complete_df.head()
```

```python
#students DF 
students_complete_df.head()
```

```python
# Combine the data into a single dataset.  
school_df = pd.DataFrame(schools_complete_df, columns=["School ID" , "school_name" , "type" , "size" , "budget"])
school_df

students_df=pd.DataFrame(students_complete_df, columns=["Student ID" , "student_name" , "gender" , "grade" , "school_name" , "reading_score" , "math_score"])
students_df

merged_df=pd.merge(school_df , students_df, on="school_name")
merged_df
```

```python
#rename columns
merged_df=merged_df.rename(columns={"school_name":"School Name" , "type":"School Type", "size":"Size" , "budget": "Budget", "student_name":"Student Name", "gender":"Gender", "grade": "Grade", "reading_score": "Reading Score", "math_score": "Math Score"})
merged_df.head()
```

```python
#Calculate the total number of schools 
schools_total=merged_df["School Name"].unique()
schools_total=len(schools_total)
schools_total
```

```python
#Calculate the total number of students
total_students=merged_df["Student ID"].count()
total_students
```

```python
#Calculate the total budget - #make it into a dollar amounttotal_budget=merged_df["Budget"].sum()
total_budget=merged_df["Budget"].unique().sum()
total_budget
```

```python
#Calculate the average math score 
math_avg=merged_df["Math Score"].mean()
math_avg
```

```python
#Calculate the average reading score
reading_avg=merged_df["Reading Score"].mean()
reading_avg
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_passing=merged_df.loc[merged_df["Math Score"]>=70]["Student ID"].count()/total_students*100
math_passing
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_passing=merged_df.loc[merged_df["Reading Score"]>=70]["Student ID"].count()/total_students*100
reading_passing
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing=(math_passing+reading_passing)/2
overall_passing
```

```python
#Create a dataframe to hold the above results
summary_df=pd.DataFrame({"Total Number of Schools": [schools_total],
                        "Total Number of Students": [total_students],
                        "Total Budget" : [total_budget],
                        "Average Math Score": [math_avg],
                        "Average Reading Score" : [reading_avg],
                        "Percentage of Students Passing Math": [math_passing],
                        "Percentage of Students Passing Reading" : [reading_passing],
                        "Percentage of Students Overall Passing":[overall_passing]}) 
summary_df["Total Budget"]=summary_df["Total Budget"].map("${:,.2f}".format)

summary_df
```

```python
#School Name
school_type=schools_complete_df.set_index("school_name")["type"]
school_type

#Total Students
students_total=merged_df["School Name"].value_counts()
students_total

#Total School Budget
total_school_budget=merged_df.groupby(["School Name"]).mean()["Budget"]
total_school_budget

#Per Student Budget
per_student=total_school_budget/students_total
per_student

#Average Math Score
avg_math_score=merged_df.groupby(["School Name"]).mean()["Math Score"]
avg_math_score

#Average Reading Score
avg_reading_score=merged_df.groupby(["School Name"]).mean()["Reading Score"]
avg_reading_score

#% Passing Math
passing_math=merged_df[merged_df["Math Score"]>=70]
percent_pass_math=passing_math.groupby(["School Name"]).count()["Student Name"]/students_total*100
percent_pass_math

#% Passing Reading
passing_reading=merged_df[merged_df["Reading Score"]>=70]
percent_pass_reading=passing_reading.groupby(["School Name"]).count()["Student Name"]/students_total*100
percent_pass_reading

#% Overall Passing (The percentage of students that passed math and reading.)
overall_passing=merged_df[(merged_df["Math Score"]>=70)& (merged_df["Reading Score"]>=70)]
overall_percent=overall_passing.groupby(["School Name"]).count()["Student Name"]/students_total*100
overall_percent
```

```python
school_summary_df=pd.DataFrame({"School Type": school_type, 
                                "Total Students": students_total, 
                                "Total School Budget": total_school_budget,
                               "Per Student Budget": per_student,
                               "Average Math Score": avg_math_score,
                               "Average Reading Score" : avg_reading_score,
                               "% Passing Math": percent_pass_math,
                               "% Passing Reading" : percent_pass_reading,
                               "% Overall Passing": overall_percent})

school_summary_df
```

```python
#top Performing Schools (By % Overall Passing

top_performing=school_summary_df.sort_values("% Overall Passing", ascending=False)
top_performing

bottom_performing=school_summary_df.sort_values("% Overall Passing", ascending=True)
bottom_performing
```
